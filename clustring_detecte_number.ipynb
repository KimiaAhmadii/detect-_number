```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

# بارگذاری دیتاست digits
digits = datasets.load_digits()

# ویژگی‌ها و داده‌ها
x = digits.data

# تقسیم‌بندی داده‌ها به آموزش و تست (بدون نیاز به y چون برچسب‌ها مهم نیستند)
X_train, X_test = train_test_split(x, test_size=0.2, random_state=42)

# مدل خوشه‌بندی K-Means با 10 خوشه
kmeans = KMeans(n_clusters=10, random_state=42)

# آموزش مدل روی داده‌های آموزشی
kmeans.fit(X_train)

# پیش‌بینی خوشه‌ها برای داده‌های تست
clusters = kmeans.predict(X_test)

# نمایش نتایج پیش‌بینی خوشه‌ها
print("خوشه‌های پیش‌بینی شده برای داده‌های تست:", clusters)

# بصری‌سازی خوشه‌های داده‌های تست
plt.scatter(X_test[:, 0], X_test[:, 1], c=clusters, cmap='viridis')
plt.title('خوشه‌بندی داده‌های تست با K-Means')
plt.show()
```
